#### Cleanup `Name` that contain "Mega".

If you have checked out the pokemon names carefully enough, you should have found there are junk texts in the pokemon names which contain "Mega". We want to clean up the pokemon names. For instance, "VenusaurMega Venusaur" should be "Mega Venusaur", and "CharizardMega Charizard X" should be "Mega Charizard X".

```python
# your code here
# clean up extract lower if mega in x x=x.split return x of 1 

#pokemon['name'] = pokemon['name'].str.replace('VenusaurMega', 'Mega')
# print(set(pokemon['name']))

# test transformed data
pokemon.head(10)
```

```python
# your code here
# pokemon.columns = pokemon.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')


def clean(x):
    x = x.lower()
    if 'mega' in x:
        x = x.split('mega')
        return ('mega' + " " + x[0])
    else:
        return(x)

pokemon['New Name']= list(map(clean,pokemon['name']))
pokemon.head()
pokemon.drop('name', axis=1)
```

```python
def combo(x,y):
    if type(x)==str and type(y)==str:
        return(x+"-"+y)
    elif type(x)==str and type(y)==float:
        return(x)
    elif type(y)==str and type(x)==float:
        return(y)  
pokemon['Combo Type'] = list(map(combo,pokemon['type_1'], pokemon['type_2']))
pokemon.head()
pokemon.drop('name', axis=1)
```
