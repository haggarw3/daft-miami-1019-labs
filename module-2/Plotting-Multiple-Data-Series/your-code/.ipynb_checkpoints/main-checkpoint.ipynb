# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

```python
# check for dtypes, null, duplicates. do basic exploratory analysis

data.dtypes
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
data_bar = data[["RetailSales", "RetailTransfers", "WarehouseSales", "ItemType"]]
data_total = data_bar.groupby("ItemType").agg("sum")
data_total
```

```python
data_total.plot.bar(figsize = (5,8), alpha = 1)
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
# create TotalSales column based on 'RetailSales' & 'WarehouseSales'

data['TotalSales'] = data[["RetailSales", "WarehouseSales"]].sum(axis=1)
data
```

```python
# sort TotalSales descending

sales_mix = data[["Supplier", "RetailSales", "WarehouseSales", "TotalSales"]] 
new_df = sales_mix.sort_values('TotalSales', ascending=[False])
new_df
```

```python
sales_mix_total = new_df.groupby("Supplier").agg("sum")
TopTen = sales_mix_total.sort_values('TotalSales', ascending=[False])
TopTen.head(10)
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
multi_line = data[["Month", "RetailSales", "RetailTransfers", "WarehouseSales"]]
multi_line_mean = multi_line.groupby("Month").agg("mean")
multi_line_mean = multi_line_mean.reset_index()
# multi_line_mean.columns # names dataframe's columns
multi_line_mean
```

```python
multi_line_mean.plot(x='Month', y=["RetailSales", "RetailTransfers", "WarehouseSales"], figsize=(20,10))
```

## 4. Plot the same information as above but as a bar chart.

```python
multi_line_mean.plot.bar(figsize = (20,20), alpha = 1)
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
multi_RetailSales = data[["Month", "Year", "RetailSales", "ItemType"]]
multi_line_sum = multi_RetailSales.groupby(['Year','Month',"ItemType"]).agg("sum")
multi_line_sum
```

```python
# based on year: 2017

piv = multi_RetailSales.pivot_table(index=['Year','Month'], columns='ItemType', values='RetailSales').reset_index()
piv = piv[piv['Year']==2017]
piv
```

```python
# to copy paste column names to plot

piv.columns
```

```python
piv.plot(x='Month', y = ['BEER', 'DUNNAGE', 'KEGS', 'LIQUOR', 'NON-ALCOHOL','REF', 'STR_SUPPLIES', 'WINE'])
```

```python
# based on year: 2018

piv2 = multi_RetailSales.pivot_table(index=['Year','Month'], columns='ItemType', values='RetailSales').reset_index()
piv2 = piv2[piv2['Year']==2018]
piv2
```

```python
piv2.plot(x='Month', y = ['BEER', 'DUNNAGE', 'KEGS', 'LIQUOR', 'NON-ALCOHOL','REF', 'STR_SUPPLIES', 'WINE'])
```

## 6. Plot the same information as above but as a bar chart.

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
import matplotlib.pyplot as plt

plt.scatter(data['RetailSales'], data['RetailTransfers'])
plt.show() 
```

```python
sns.lmplot( x="RetailSales", y="RetailTransfers", data=data, fit_reg=False, hue='ItemType', legend=True)
```
