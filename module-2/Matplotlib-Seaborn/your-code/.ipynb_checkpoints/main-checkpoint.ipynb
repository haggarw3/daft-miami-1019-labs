#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
#fig = plt.figure()
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y)
ax1.set_title('X vs Y linear Function')
ax2.plot(x, z)
ax2.set_title('X vs Z Cuadratic Function')
plt.show()
```

```python
def plotting(ax, x, y, pmtrs):
    return ax.plot(x, y, **pmtrs)
fig, [ax1, ax2] = plt.subplots(1, 2)
plotting(ax1, x, y, {'marker': '_'})
plotting(ax2, x, z, {'marker': '.'})
plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=[20.0, 10.0], facecolor= (0.75, 0.45, 0.25))
ax1.plot(x, y)
ax1.set_title('X vs Y', fontsize=50, backgroundcolor=(0.5, 0.5, 0.5))
ax2.plot(x, z)
ax2.set_title('X vs Z', fontsize = 30, loc='right', rotation=20, fontstyle='italic')
plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
w = np.arange(0.1, 5, 0.1)
quadratic_values = np.square(w)
exp_values = np.exp(w)

fig, ax1 = plt.subplots()

ax1.plot(w, quadratic_values, color='b')
ax1.tick_params(axis='y')

ax2 = ax1.twinx()

ax2.plot(w, exp_values, color='y')
ax2.set_yscale('log')
ax2.tick_params(axis='y')

plt.show()
```

```python
fig, [axs1, axs2, axs3, axs4] = plt.subplots(1, 4, figsize=[15.0, 5.0])
axs1.plot(w, np.square(w))
axs1.set_title('Quadratic Function')
axs2.plot(w, np.exp(w))
axs2.set_title('Exponential Function')
axs3.plot(w, np.square(w), color='r')
axs3.set_title('Quadratic Function (log scale)')
axs3.set_yscale('log')
axs4.plot(w, np.exp(w), color='r')
axs4.set_title('Exponential Function (log scale)')
axs4.set_yscale('log')
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
fig, [axs1, axs2, axs3, axs4] = plt.subplots(1, 4, figsize=[15.0, 5.0])
fig.suptitle('Scales')
axs1.plot(w, np.square(w), label='Quad curve')
axs1.set_title('Quadratic Function')
axs1.legend()
axs2.plot(w, np.exp(w), label='Exp curve')
axs2.set_title('Exponential Function')
axs2.legend()
axs3.plot(w, np.square(w), color='r', label='Quad curve')
axs3.set_title('Quadratic Function (log scale)')
axs3.set_yscale('log')
axs3.legend()
axs4.plot(w, np.exp(w), color='r', label='Exp curve')
axs4.set_title('Exponential Function (log scale)')
axs4.set_yscale('log')
axs4.legend(fontsize=10)
plt.show()
```
