#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

cols = ['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary', 'Steps']

df= fitbit[cols]
fig, axes= plt.subplots(figsize=(12,4),nrows=1,ncols=3)

for i in range(3):
    plt.sca(axes[i])
    plt.scatter(df.iloc[:,i], df['Steps'], alpha = 1) 
    plt.xlabel(df.iloc[:,i].name) #x label will be column name
    plt.ylabel('Steps') #y label will be steps
```
