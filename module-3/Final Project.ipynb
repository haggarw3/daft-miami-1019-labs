```python
zillow_2016 = zillow_2016.drop(columns=['architecturalstyletypeid','buildingclasstypeid','decktypeid','finishedfloor1squarefeet', 
                          'finishedsquarefeet13', 'finishedsquarefeet15', 'finishedsquarefeet50', 'finishedsquarefeet6',
                          'poolsizesum','storytypeid', 'yardbuildingsqft17','yardbuildingsqft26', 'taxdelinquencyyear'])
```

```python
zillow_2016.isna().sum()
```

```python
zillow_2016['fireplaceflag'].fillna("False", inplace = True) 
zillow_2016['hashottuborspa'].fillna("False", inplace = True) 
zillow_2016['airconditioningtypeid'].fillna(5.0, inplace = True) 
zillow_2016['basementsqft'].fillna(0., inplace = True) 
zillow_2016['bathroomcnt'].fillna(0., inplace = True) 
zillow_2016['bedroomcnt'].fillna(0., inplace = True) 
zillow_2016['fireplacecnt'].fillna(0., inplace = True) 
zillow_2016['fullbathcnt'].fillna(0., inplace = True) 
zillow_2016['calculatedbathnbr'].fillna(0., inplace = True) 
zillow_2016['garagecarcnt'].fillna(0., inplace = True) 
zillow_2016['heatingorsystemtypeid'].fillna(13., inplace = True) 
zillow_2016['poolcnt'].fillna(0., inplace = True)
zillow_2016['pooltypeid10'].fillna(0., inplace = True)
zillow_2016['pooltypeid2'].fillna(0., inplace = True)
zillow_2016['pooltypeid7'].fillna(0., inplace = True)
zillow_2016['roomcnt'].fillna(0., inplace = True)
zillow_2016['threequarterbathnbr'].fillna(0., inplace = True)
zillow_2016['typeconstructiontypeid'].fillna(0., inplace = True)
zillow_2016['numberofstories'].fillna(0., inplace = True)
zillow_2016['taxdelinquencyflag'].fillna('N', inplace = True)
zillow_2016['garagetotalsqft'].fillna(0., inplace = True)
```

```python
zillow_2016 = zillow_2016.dropna(subset=['buildingqualitytypeid'])
zillow_2016 = zillow_2016.dropna(subset=['calculatedfinishedsquarefeet'])
zillow_2016 = zillow_2016.dropna(subset=['finishedsquarefeet12'])
zillow_2016 = zillow_2016.dropna(subset=['fips'])
zillow_2016 = zillow_2016.dropna(subset=['lotsizesquarefeet'])
zillow_2016 = zillow_2016.dropna(subset=['propertycountylandusecode'])
zillow_2016 = zillow_2016.dropna(subset=['propertylandusetypeid'])
zillow_2016 = zillow_2016.dropna(subset=['regionidcity'])
zillow_2016 = zillow_2016.dropna(subset=['regionidcounty'])
zillow_2016 = zillow_2016.dropna(subset=['regionidneighborhood'])
zillow_2016 = zillow_2016.dropna(subset=['regionidzip'])
zillow_2016 = zillow_2016.dropna(subset=['bathroomcnt'])
zillow_2016 = zillow_2016.dropna(subset=['taxvaluedollarcnt'])
zillow_2016 = zillow_2016.dropna(subset=['yearbuilt'])
zillow_2016 = zillow_2016.dropna(subset=['structuretaxvaluedollarcnt'])
zillow_2016 = zillow_2016.dropna(subset=['taxamount'])
zillow_2016 = zillow_2016.dropna(subset=['landtaxvaluedollarcnt'])
zillow_2016 = zillow_2016.dropna(subset=['propertyzoningdesc'])
zillow_2016 = zillow_2016.dropna(subset=['unitcnt'])
zillow_2016 = zillow_2016.dropna(subset=['censustractandblock'])
```
